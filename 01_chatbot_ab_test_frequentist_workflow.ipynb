## What are the goals of an A/B test and how is it set up?

The goal of our A/B test is to establish whether the chatbot provides a benefit to **users** and to the **company**. We all hoped this would be the case, but unless we conduct a proper A/B test, we cannot know for sure.

To maintain our intellectual integrity and pave the way for the rest of the company, we set up the A/B test in the following manner:

- for a random **50% of users**, the new chatbot was disabled. Those users were part of the **Control** group, or **A** group.
- for a random **50% of users**, the new chatbot was enabled. Those users were part of the **Test** group, or **B** group.

It is essential for users to be **randomly assigned** to control and test groups. This helps ensure that both groups have a similar distribution of user characteristics. If we follow a non-random system when assigning users, the groups may stop being comparable. If that happens, the difference we observe in outcomes might not be due to the chatbot at all, but due to the underlying differences that we introduced ourselves.

In addition, users did not know that the experiment was happening or which group they were assigned to. This helps ensure that behaviour is not biased by knowledge of the experiment. Knowing that one is part of an experiment can make people behave in strange or unusual ways. See the [Hawthorne effect](https://en.wikipedia.org/wiki/Hawthorne_effect).

> If users know they are part of an experiment, we may stop observing their **normal behaviour**. In that case, our conclusions may no longer apply to the real world.

In medical research, this type of experiment is called a [randomised single-blind controlled experiment](https://en.wikipedia.org/wiki/Blinded_experiment).

We are not unique in conducting such experiments. In fact, almost every company with a website is likely to be conducting A/B tests. The goal is to detect which features work best for users while also making the company more profitable. You have probably participated in thousands of A/B tests just by browsing your favourite websites without ever knowing. Being a test subject in an experiment has never been more fun!


## Simple Example for Confounding

A trucking company compares the fuel economy of trucks from two manufacturers (“A” and “B”) by measuring miles per gallon over one month. They find that A trucks appear more fuel-efficient. However, A trucks are more often assigned highway routes while B trucks are more often assigned city routes. Here, truck make is the independent variable, MPG is the dependent variable, and route type (or proportion of city driving) is the confounder. Because route type affects MPG and the route type differs across truck make, it confounds the comparison. Thus the observed difference likely reflects highway vs. city driving rather than truck make.

## A/B tests vs Observational studies

To evaluate the impact our chatbot had, we could have done something quite different:

1. Record the most important metrics **before** the rollout of the chatbot.
2. Roll out the chatbot to **100%** of users.
3. Record the most important metrics **after** the rollout.
4. Compare the metrics from before and after the rollout and decide whether the chatbot is helping.

This approach is called an **observational study** and is frequently used when a real experiment is not possible.

The major problem with observational studies is that **causality** is very difficult to prove. The main challenge comes from a phenomenon called [confounding](https://en.wikipedia.org/wiki/Confounding).

Let’s recap what we are trying to achieve by running an experiment. We want data that helps us evaluate whether the chatbot benefits both users and the company. If we released the bot to 100% of customers and later saw that user experience and profits had increased, could we objectively claim that this happened *because* of the chatbot?

The temptation to say yes is huge. We worked hard to create the chatbot, so of course we want it to be successful. But many other things might have happened around the same time as the release, for example:

- a successful marketing campaign was launched
- Agentur für Arbeit got additional financing and made it easier to secure funding
- Germany entered a phase of rapid economic growth and businesses started hiring more than before

The list goes on and on.

Any one of the reasons above could have been the reason why we have seen the increase in the important metrics rather than the chatbot.
More generally the problem with observational studies is that you cannot exclude influence of other **confounders** on the outcome of your experiments. In professional language you can't **control** for those other confounders. 

If you're not convinced by this simple explanation, I would like to refer you to [Refuted observation studies](https://experimentguide.com/refuted_observational_studies/) and the [The book of Why](https://en.wikipedia.org/wiki/The_Book_of_Why).

### How does an A/B test deal with a problem of confounding?

> **Confounding** means that multiple things may be changing at the same time, so we can no longer tell what caused the observed effect.

If you recall when splitting users into test and control groups we were using a random approach, because we are trying to make sure that test and control groups are identical or at least extremely similar to each other. In fact what we are aiming to achieve is:

*To make test group different from control group in 1 thing only - presence or absence of the chatbot*. 

By doing this we are not trying to remove the confounders from the experiment. We are helpless against the confounders, but we are not helpless in creating extremely similar groups that have 1 difference only - the chatbot we are trying to test. Under those conditions : when the groups are identical to each other in everything except presence or absence of a chatbot, if we observe difference between the groups, we can conclude that it's *because of the chatbot*. In other words - A/B tests are the only reliable tool available to us that can help us infer **causality**.

We together with the rest of the tech industry were convinced that A/B test is the right way to go and we had the technology to make it happen. Hence we ran an A/B tests.

### HiPPOs 🦛

There is another way to make rollout decisions: the **HiPPO** system.

> **HiPPO** stands for **Highest Paid Person in the meeting**.

This system is rather simple: the highest-paid person in the room tells everyone what to do.

- **Advantage:** we can do things *fast*.
- **Disadvantage:** we *rarely know what effect those fast changes actually had*.

This system can appear in at least two scenarios:

- a very young company that simply does not yet have the resources to do A/B testing
- an established company that is not **data driven**, but driven by the gut feelings of various HiPPOs

HiPPOs are part of the real world, and you will undoubtedly encounter them in your career. To encourage you: HiPPOs mostly eat plant-based food, so you should not be too afraid of their large open mouths. [More on HiPPOs](https://jeffgothelf.com/blog/highest-paid-persons-opinion/)

Did we have HiPPOs of our own when rolling out our first A/B test? Of course we did. A week after our A/B test had begun, one of our own HiPPOs insisted that a WhatsApp button enabling users to contact the admissions team needed to be added to the website. The reasoning was: *“All our competitors do that, therefore we also must do that.”*

The button was rolled out *fast* to 100% of users, without an A/B test.

Nobody will ever know what the effect of that release was. We will also never know what effect it had on the ongoing chatbot A/B test. Both functionalities allow users to ask questions in written form, so rolling out a WhatsApp button in the middle of an ongoing A/B test clearly had the potential to influence the test results.

HiPPOs usually want to do what is best, but they do not know any better. As data professionals, it is our duty to educate them. And to be fair, HiPPOs might sometimes be right. We just rarely get proper evidence of that.


## Website tracking

To understand the A/B test results, we first need to discuss **how users were tracked**.

Users were tracked using **Google Analytics** (`ga`) - a system that registers **events** triggered when a user interacts with the website. In our case, the following events were configured:

- `experiment_viewed_ga_first_timestamp` - the event registered when a user is assigned to either the test or control group. In other words, every user who was part of the experiment will have this timestamp set.
- `experiment_ga` - which experiment the user was part of. The expected value here is `chatbot`, but there could have been other A/B tests before.
- `variation_ga` - `false` for chatbot disabled and `true` for chatbot enabled.
- `device_category_ga` - which device the user used to access the website.
- `application_started_first_timestamp` - the event registered the first time a user clicks the **Start now** button to open the application form.
- `form_interaction_first_timestamp` - the event registered the first time a user enters some details into the application form.
- `application_first_timestamp` - the event registered the first time a user submits the application form.

### Customer relationship management (CRM) system tracking

Once a user submits their application form, they are registered in the customer relationship management system **HubSpot**. The following tracking events were configured:

- `lead_first_timestamp` - the event registered the first time a user submits the application form. In theory, it should be exactly the same as `application_first_timestamp` from the website tracking section. In practice, there are discrepancies between the two. This is normal: two different systems are involved, and various bugs can prevent them from being perfectly in sync.
- `interview_first_timestamp` - the event registered the first time a user is scheduled for an interview with the admissions team.
- `admitted_first_timestamp` - the event registered the first time a user is admitted to the bootcamp. This happens if the interview goes well and the user passes the admission test.
- `financing_confirmed_first_timestamp` - the event registered the first time a user secures funding for the bootcamp.

These discrepancies are very common in the real world, so it is good to get used to them.

You can probably see where this is heading. When tables are connected through **1-n** or **n-n** relationships, the number of rows in the final dataset can grow very quickly. The more joins you do, the more rows get multiplied.

While **150,000 rows** is still manageable on an average computer, datasets with millions of rows can clog memory and make a laptop extremely slow.

> It often makes sense to create a convenient dataset with the minimum necessary number of rows **at the data-selection stage**, rather than only dealing with the problem later.

So the question becomes: **how?**


## Simplifying assumptions to make the dataset smaller

The main question we need to answer for each event is this:

> If a user had multiple interactions with the application form or with any other event, is that information vital enough that it needs to be present in the final dataset?

For the sake of evaluating the result of this A/B test, is it important to know:

1. that the user clicked on the application form page
2. that the user clicked on the application form page **17 times**
3. both

A reasonable simplifying assumption is the following:

> *We care whether the user performed a certain action on the website or not, and we do not care how many times the user did it.*

It is true that some information will be lost, but for the sake of the current analysis we can probably live with that. If you disagree, you are welcome to conduct the analysis with another simplifying assumption, or without making any simplifying assumptions at all. Just be clear about it to yourself and to others.

If this assumption seems reasonable, we can aim for a final row count of **67,000** by selecting one instance of each multiple event for each user. For example, we could look at the *first* time the user clicked on the application form, the *first* time they interacted with it, and the *first* time they submitted it. Or we could look at the *median* interaction or the *last* interaction instead.

This is exactly the simplifying approach reflected in the prepared dataset used below.

# Evaluating the A/B test

We have finally reached the stage where we can start evaluating the results of our A/B test.

Let’s first state the business question we are trying to answer:

> **Did the chatbot improve user experience and did it help the company make more money?**

By the very way the question is phrased, we can already see that we are trying to achieve **two separate goals**:

1. **Improve user experience**
2. **Make more money**

In an ideal world, those two goals would be perfectly correlated: better user experience would always lead to making more money. But the real world is more complicated than that.

It is entirely possible, and actually not that uncommon, that better user experience *does not* result in making more money. Sometimes it can even end up making us lose money.

## Better user experience

It is a sad but empirical fact that we are often very bad at understanding what users actually want. If we were not, A/B testing would be completely unnecessary. We would simply imagine how a feature should work and it would work perfectly from the initial rollout.

Life would be easy and simple. On the other hand, there would also be no need for data professionals to help design and evaluate A/B tests.

The first thing to note is that there is likely to be a high degree of diversity among our users. When we roll out a new product, some users will hate it, some will be indifferent, and some will love it. Better user experience usually means **what works best for the majority**.

More often than not, the initial rollout of a new feature does **not** improve user experience. This is true despite the best efforts of everyone involved. If you do not believe me, please get yourself a copy of [Trustworthy online controlled experiments](https://www.cambridge.org/core/books/trustworthy-online-controlled-experiments/D97B26382EB0EB2DC2019A7A7B518F59).


## Making more money

This is probably the ultimate goal of every business. The only way for a business to survive and stay competitive in a fast-changing world is to innovate.

It is probably possible to have a business that does not grow and has stable, unchanging income - a local bakery comes to mind. However, it is unlikely that you will be working for such a business after completing a data bootcamp.

Fundamentally, if a business stops innovating, it starts losing out to competitors, stops being able to pay for itself, and eventually disappears. Those are the tough laws of life in a free-market economy.

All that said, the primary goal of most product rollouts is to **make more money**, even if it means making the user experience worse. At the very least, we want to see that the new product is **not causing us to lose money**.

If we do see that a recent rollout is losing money, we should admit failure, roll back the release, and go back to the drawing board to rethink what the product is supposed to do and why it is helpful. Colleagues from **UX/UI** and **product** should be able to help us with this.

## Why do we need statistics?

We have computed the `application_started` conversion rate, and we could easily compute the other conversion rates too. At first glance, it may seem like the job is done.

Sadly, it is not. But do not fret, we are almost there.

The problem with merely computing proportions or conversion rates is this: how can we be sure that the users who interacted with our experiment are, statistically speaking, **representative** users?

Perhaps we were extremely unlucky and happened to get users who passionately hate chatbots. If we ran the experiment for a few more weeks, maybe we would get more chatbot-positive users and the overall outcome would change.

To make this more concrete, imagine we had only **10 people** in the control version and **10 people** in the test version. If 3 people converted in the control group and 4 converted in the test group, we would observe:

- **30%** conversion in the control group
- **40%** conversion in the test group

Would that be enough to declare success?

The naive answer is **yes**, and you should expect to hear it often in business life. The educated answer is **no**.

> What we need is a procedure that tells us whether the result we are observing is **statistically significant** and whether the test is properly **powered**.

That way, we can judge whether the observed effect is likely to be real, rather than just random luck or bad luck.


## A/B test evaluation function

The function below is probably the most difficult thing to understand in this whole notebook.

> Before diving into its details, let’s first look at the **output** and the **plots** it produces in the next section.


````python
def evaluate_ab_test_with_plots_and_power(
    conversions_a,
    trials_a,
    conversions_b,
    trials_b,
    significance_level=0.05,
    power_target=0.80,
    title_binary="Density Plot of Binary Outcomes",
    title_conversion="Density Plot of Conversion Rates",
):
    """
    Evaluates an A/B test, performs a power check, and generates plots.

    Args:
        conversions_a (int): Number of conversions in the control group.
        trials_a (int): Total number of trials in the control group.
        conversions_b (int): Number of conversions in the treatment group.
        trials_b (int): Total number of trials in the treatment group.
        significance_level (float): The alpha level for the test.
        power_target (float): The desired statistical power level.
        title_binary (str): Title for the binary outcomes plot.
        title_conversion (str): Title for the conversion rates plot.

    Returns:
        dict: A dictionary containing test results including a power check.
    """
    # Calculate conversion rates
    conversion_rate_a = conversions_a / trials_a
    conversion_rate_b = conversions_b / trials_b

    # --- Statistical Significance (P-Value & Z-Score) ---
    pooled_proportion = (conversions_a + conversions_b) / (trials_a + trials_b)
    standard_error = np.sqrt(
        pooled_proportion * (1 - pooled_proportion) * (1 / trials_a + 1 / trials_b)
    )
    z_score = (conversion_rate_b - conversion_rate_a) / standard_error
    p_value = (1 - norm.cdf(abs(z_score))) * 2
    is_significant = p_value < significance_level

    # --- Statistical Power Check ---
    # The minimum detectable effect (MDE) is the observed difference
    observed_effect_size = conversion_rate_b - conversion_rate_a

    # Check for a zero observed effect to avoid division by zero.
    if observed_effect_size == 0:
        required_sample_size = float("inf")
        is_powered = False
    else:
        # Get Z-scores for significance and power targets
        Z_alpha = norm.ppf(1 - significance_level / 2)
        Z_beta = norm.ppf(power_target)

        # Calculate the required sample size for the observed effect
        numerator = (Z_alpha + Z_beta) ** 2 * (
            conversion_rate_a * (1 - conversion_rate_a)
            + conversion_rate_b * (1 - conversion_rate_b)
        )
        denominator = observed_effect_size**2
        required_sample_size = np.ceil(numerator / denominator)

        # Compare actual sample size to required sample size
        is_powered = (trials_a >= required_sample_size) and (
            trials_b >= required_sample_size
        )

    # --- Plotting ---
    np.random.seed(0)  # For reproducibility
    dist_a = np.random.binomial(1, conversion_rate_a, trials_a)
    dist_b = np.random.binomial(1, conversion_rate_b, trials_b)

    sample_proportions_a = np.random.normal(
        conversion_rate_a,
        np.sqrt(conversion_rate_a * (1 - conversion_rate_a) / trials_a),
        1000,
    )
    sample_proportions_b = np.random.normal(
        conversion_rate_b,
        np.sqrt(conversion_rate_b * (1 - conversion_rate_b) / trials_b),
        1000,
    )

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.kdeplot(dist_a, label="Control Group (A)", fill=True)
    sns.kdeplot(dist_b, label="Treatment Group (B)", fill=True)
    plt.title(title_binary)
    plt.xlabel("Outcome")
    plt.ylabel("Density")
    plt.legend()

    plt.subplot(1, 2, 2)
    sns.kdeplot(sample_proportions_a, label="Control Group (A)", fill=True)
    sns.kdeplot(sample_proportions_b, label="Treatment Group (B)", fill=True)
    plt.title(title_conversion)
    plt.xlabel("Conversion Rate")
    plt.ylabel("Density")
    plt.legend()

    plt.tight_layout()
    plt.show()

    return {
        "conversion_rate_a": conversion_rate_a,
        "conversion_rate_b": conversion_rate_b,
        "z_score": z_score,
        "p_value": p_value,
        "is_significant": is_significant,
        "required_sample_size": required_sample_size,
        "is_powered": is_powered,
    }
````

### Z-scores

A **Z-score** is a statistical measurement that describes a value’s relationship to the mean of a group of values. It is measured in terms of **standard deviations** from the mean.

- If a Z-score is `0`, the data point is exactly on the mean.
- A **positive** Z-score indicates that the data point is above the mean.
- A **negative** Z-score indicates that the data point is below the mean.

Z-scores are a form of **standardization** that allows you to compare data points from different datasets. For example, you can compare a student's test score in one class to a student's score in another class, even if the tests had different maximum scores and different average scores.

#### **Formula and Calculation**

The formula for calculating a Z-score is:

$Z = \frac{(x - \mu)}{\sigma}$

Where:

- $Z$ is the Z-score
- $x$ is the raw data point you want to standardize
- $\mu$ (mu) is the mean of the population
- $\sigma$ (sigma) is the standard deviation of the population

Let’s consider a simple example: a group of students takes a math test. The average score ($\mu$) is `75`, and the standard deviation ($\sigma$) is `10`.

- A student who scores `85` has a Z-score of $Z = \frac{(85 - 75)}{10} = 1$. This means their score is **one standard deviation above the mean**.
- A student who scores `65` has a Z-score of $Z = \frac{(65 - 75)}{10} = -1$. This means their score is **one standard deviation below the mean**.
- A student who scores `75` has a Z-score of $Z = \frac{(75 - 75)}{10} = 0$. Their score is **exactly at the mean**.


````python
def plot_z_score_distribution(mean=0, std_dev=1, num_samples=5000):
    """
    Generates and plots a normal distribution with Z-scores on the x-axis.

    Args:
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
        num_samples (int): The number of samples to generate.
    """
    # Generate a normally distributed dataset
    data = np.random.normal(mean, std_dev, num_samples)

    # Calculate Z-scores for each data point
    _ = (data - mean) / std_dev

    # Set up the plot
    plt.figure(figsize=(10, 6))

    # Create the density plot
    sns.kdeplot(x=data, fill=True, color="skyblue")

    # Set up Z-score ticks and labels
    # Define key Z-score values to be displayed on the axis
    z_score_labels = np.array([-3, -2, -1, 0, 1, 2, 3])

    # Convert Z-scores back to raw data points for setting tick locations
    raw_data_ticks = z_score_labels * std_dev + mean

    # Set the tick locations on the x-axis
    plt.xticks(raw_data_ticks)

    # Set the tick labels to be the Z-scores
    ax = plt.gca()
    formatter = FuncFormatter(lambda x, pos: f"{round((x - mean) / std_dev):.0f}")
    ax.xaxis.set_major_formatter(formatter)

    # Add vertical lines at key Z-score points for visual aid
    for z in z_score_labels:
        plt.axvline(
            x=z * std_dev + mean, color="gray", linestyle="--", linewidth=1, alpha=0.7
        )

    # Add labels and title
    plt.title("Normal Distribution with Z-Scores", fontsize=16)
    plt.xlabel("Z-Score", fontsize=12)
    plt.ylabel("Density", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.6)

    plt.show()


# Example usage of the function with a custom mean and standard deviation
plot_z_score_distribution(mean=100, std_dev=15)

# Example usage with the default mean and standard deviation (0 and 1)
# plot_z_score_distribution()
````

#### **Uses of Z-Scores**

### Null Hypothesis

The null hypothesis (H₀) in A/B testing assumes that there is **no difference** between the two groups (A and B), meaning any observed difference is due to random chance. The goal is to test whether the observed difference between groups is more extreme than what would reasonably occur by chance alone.

If the data shows that the difference is statistically significant, meaning the probability of observing such a difference by chance (the p-value) is below a predefined significance level such as `p < 0.05`, we can **reject** the null hypothesis. If not, we fail to reject H₀, but that does not prove the null hypothesis is true. It only means there is insufficient evidence against it.

We can conclude that the groups differ in a meaningful way only when the evidence is strong enough. If the difference is not significant, we do not yet have enough evidence to claim a true difference.

Statisticians use slightly different terminology to talk about the kinds of errors the model might be making:

| | **Predicted Positive** | **Predicted Negative** |
| :--- | :---: | :---: |
| **Actual Positive** | True Positive (TP) | False Negative (FN) - Type 2 Error |
| **Actual Negative** | False Positive (FP) - Type 1 Error | True Negative (TN) |

A **Type I error** occurs when a test result or conclusion incorrectly indicates that a condition is present when it is not. This is also known as a **false positive**. In statistical terms, it is the **error of rejecting a true null hypothesis**.

* Example: A medical test says a patient has a disease, but they are actually healthy.
* In hypothesis testing: the null hypothesis states there is no effect or no difference, for example that a new drug has no effect. A Type I error happens if you conclude that the drug works when, in reality, it does not.

A **Type II error** occurs when a test result or conclusion incorrectly indicates that a condition is absent when it is actually present. This is also known as a **false negative**. In statistical terms, it is the **error of failing to reject a false null hypothesis**.

* Example: A medical test says a patient does not have a disease, but they are actually sick.
* In hypothesis testing: using the same drug example, a Type II error happens if you conclude that the drug does not work when, in reality, it does have an effect.

When conducting an A/B test, we have to be mindful of both types of errors. We will consider how to deal with each one in detail.

Before we do, I recommend reading [this excellent article on power and statistical significance](https://cxl.com/blog/statistical-power/?utm_source=chatgpt.com#h-what-is-statistical-power).

### Statistical power (Type 2 Error / False Negatives)

A power test measures the ability of an A/B test to detect a true effect or difference between groups if one exists. The power of a test is the probability of correctly rejecting the null hypothesis when the alternative hypothesis is true, meaning it avoids a Type II error. Typically, a power of 80% or higher is desired.

To compute whether the test is properly **powered**, we need to calculate the minimum detectable effect (MDE):

$$MDE = (Z_{1-\alpha/2} + Z_{1-\beta}) \times \sqrt{\frac{p_1(1-p_1)}{n_1} + \frac{p_2(1-p_2)}{n_2}}$$

### **Explanation of Variables**

* **MDE:** The **minimum detectable effect**, expressed as the smallest difference in proportions that a test is likely to find statistically significant
* **$Z_{1-\alpha/2}$:** The Z-score corresponding to your desired significance level ($\alpha$)
* **$Z_{1-\beta}$:** The Z-score corresponding to your desired statistical power ($1-\beta$)
* **$p_1$:** The baseline conversion rate of the control group
* **$p_2$:** The expected conversion rate of the test group
* **$n_1$ and $n_2$:** The sample sizes of the control and test groups

In essence, the MDE is determined by a **balance of desired risk and available resources**. A lower MDE, meaning you can detect smaller effects, requires a larger sample size.

Run the next cell to calculate the observed effect size, the required sample size, and whether this worked example is adequately powered.


````python
# Statistical power calculations for the walkthrough example
walkthrough_observed_effect_size = (
    walkthrough_conversion_rate_b - walkthrough_conversion_rate_a
)

if walkthrough_observed_effect_size == 0:
    walkthrough_required_sample_size = float("inf")
    walkthrough_is_powered = False
    walkthrough_Z_alpha = None
    walkthrough_Z_beta = None
else:
    walkthrough_Z_alpha = norm.ppf(1 - walkthrough_significance_level / 2)
    walkthrough_Z_beta = norm.ppf(walkthrough_power_target)
    walkthrough_required_sample_size = np.ceil(
        (walkthrough_Z_alpha + walkthrough_Z_beta) ** 2
        * (
            walkthrough_conversion_rate_a * (1 - walkthrough_conversion_rate_a)
            + walkthrough_conversion_rate_b * (1 - walkthrough_conversion_rate_b)
        )
        / walkthrough_observed_effect_size**2
    )
    walkthrough_is_powered = (
        walkthrough_trials_a >= walkthrough_required_sample_size
        and walkthrough_trials_b >= walkthrough_required_sample_size
    )

{
    "observed_effect_size": walkthrough_observed_effect_size,
    "Z_alpha": walkthrough_Z_alpha,
    "Z_beta": walkthrough_Z_beta,
    "required_sample_size": walkthrough_required_sample_size,
    "is_powered": walkthrough_is_powered,
}
````

# Additional Explanation Gemini Pro for Statistacally Significant: 

Here is a clear breakdown of what statistical significance means, how to spot it in your data, and how to use it to make decisions, based on the chatbot A/B testing chapter.

### What Does "Statistically Significant" Mean?

In simple terms, **statistical significance** is a mathematical proof that the result you are seeing is real and not just a product of random chance (or "dumb luck").

Whenever you run an A/B test, you start with the **Null Hypothesis ($H_0$)**, which assumes that your new feature (the chatbot) makes absolutely no difference. If your results show a difference between the control and test groups, statistical significance tells you whether that difference is extreme enough to prove the Null Hypothesis wrong.

If a result is statistically significant, you can confidently say, *"The chatbot actually caused this change in user behavior."* If it is not significant, you have to say, *"This difference might just be random noise; we don't have enough evidence to blame or praise the chatbot."*

---

### How Do You Observe It?

In the provided Python function (`evaluate_ab_test_with_plots_and_power`), you can observe statistical significance in two distinct ways: numerically and visually.

**1. Numerically (The Math)**
When you output the dictionary of results, look for three specific metrics:

* **`p_value`:** This is the probability that the difference you observed happened by pure chance. You compare this to your **Significance Level ($\alpha$)** (usually set at 0.05, meaning you accept a 5% risk of a false positive). If your `p_value` is **lower** than 0.05, your result is statistically significant.
* **`is_significant`:** This is a simple boolean flag in the code. If it reads `True`, you have achieved statistical significance.
* **`z_score`:** This measures how many standard deviations your result is away from the expected average. Generally, a Z-score greater than **1.96** or less than **-1.96** indicates statistical significance at the 95% confidence level.

**2. Visually (The Plots)**
Look at the right-hand chart generated by the code (the "Conversion Rates Distribution").

* **Significant:** The bell curves for Group A (Control) and Group B (Treatment) are pushed far apart and barely overlap.
* **Not Significant:** The two bell curves are sitting right on top of each other or heavily overlapping, meaning the behavior of the two groups is virtually indistinguishable.

---

### How Do You Use It in A/B Testing?

Statistical significance is the ultimate referee in product development. Here is how you use it practically:

**1. To Guard Against "HiPPO" Decisions**
As the text mentions, the "Highest Paid Person's Opinion" often drives fast, gut-based changes. You use statistical significance to replace opinions with facts. If a HiPPO says, "The chatbot is working!" but your test is *not* statistically significant, you use the data to prove that any perceived success is currently just an illusion of random chance.

**2. To Make the Final Call (Win, Lose, or Wait)**

* **If it IS significant (and adequately powered):** You can safely make a business decision. In the chatbot example, the metrics for "Application Started" and "Form Interaction" were statistically significant, but the conversion rates *dropped*. Because the data was significant, the team could confidently conclude that the chatbot was actively hurting the user experience at the top of the funnel.
* **If it is NOT significant:** You **cannot trust the result yet**. For example, the "Application Submitted" metric showed a slight drop, but the p-value was 0.203 (not significant). You use this lack of significance as a warning sign to *not* make a decision yet.

**3. Pairing it with Statistical Power**
You must always use statistical significance in tandem with **Statistical Power** (avoiding false negatives). If your result is not statistically significant, you look at the `is_powered` flag. If the test is underpowered, it means you simply didn't have enough users in the test to prove anything. You use this information to tell the business: *"We need to run this test for another three weeks to get a larger sample size before we can trust the numbers."*
