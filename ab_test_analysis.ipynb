## What are the goals of an A/B test and how is it set up?
The goal of our A/B test is to establish whether the chatbot provides benefit to the users and to the company. We and everyone else hoped this would be the case, but unless we conduct a propper A/B test we can't ever know for sure. To maintain our intellectual integrity and pave the way for the rest of the company we have setup the A/B test in the following manner :


*   For random 50% of the users the new chatbot was disabled -those users were part of the 'Control' group ('A' group)
*   For random 50% of the users the new chatbot was enabled -those users were part of the 'Test' group ('B' group)

It is essential for users to be *randomly* assigned into control and test groups to ensure that both control and test groups have a similar distribution of user qualities. If we follow a system when selecting users into the Test and Control groups, we might make the groups not be similar to each other. If this happens the difference we are observing in the outcomes might not be due to the new functionality, but due to underlying differences between the test and the control group that we ourselves introduced when creating the groups. \
In addition the users didn't know the experiment was going on and which group they were assigned into to ensure that their behaviour is not biased by the knowledge that they are part of the experiment. Knowing that one is part of an experiment can make people behave in very strange, unusual ways - see [Hawthorne effect](https://en.wikipedia.org/wiki/Hawthorne_effect). If our users find out they are part of an experiment we might not observe their normal behaviour and the result of the A/B test will not be representative of our "statistically average" customer's behaviour on the website. In other words if our users know about the experiment they are part of, our conclusion might not be applicable to the world outside of our experiment. If this happens our experiment was of no use, so we got to keep it as much a secret as possible 🤫. Ideally, we never mention the experiment anywhere and our users don't know each other and can't gossip about latest things they have seen on our website either. If our users are completely oblivious of the experiment happening we can say that our experiment is reflection of the real world and that our conclusions are valid.\
In medical research this type of experiment is called ['Randomised single-blind controlled experiment'](https://en.wikipedia.org/wiki/Blinded_experiment). \
We are not inique in conducting such experiments, in fact every company with a website is likely to be conducting A/B tests. The goal of those is to detect which features work best for the user whilst making the company more profitable. You've probably unknowingly participated in thousands of A/B tests by browsing your favourite websites without ever knowing. Being a test subjects in an experiment🧪 has never been more fun!



## A/B tests vs Observational studies
To evaluate the impact our chatbot had we could have done something quite different :

1.   We record the most important metrics before the rolllout of the chatbot.
2.   We roll out the chatbot too 100% of the users.
3.   We record the most important metrics after the rollout of the chatbot.
4.   We compare the metrics before the rollout to the metrics after the rollout and decide whether the chatbot is helping or not.

This approach is called an **observation study** and is frequently used in situations where an experiment is not possible. The major problem with observational studies is that **causality** is close to impossible to prove. The main problem is with a phenomena called [confounding](https://en.wikipedia.org/wiki/Confounding). I will try to explain it in simple terms below. 

Let's recap what we are trying to achieve by running an experiment : 
We would like to get data that will help us evaluate whether the chatbot is benefiting both users and the company. If we release the bot to 100% of the customers and see that both user experience and our profits have increased, can we objectively claim that this is because of our chatbot? \
The temptation to do that is huge! We have worked so hard to create the chatbot, put our best thoughts and effort into it - of course it should work and help we are thinking to ourselves!\
The problem is that many other things might have happened around the time of our release and following it. Perhaps one or many of the following events happened around the time of the chatbot release :

*   A successful marketing campaign was launched
*   Agengur für Arbeit got additional financing and made it easier to secure funding
*   Germany entered a stage of rapid economic growth and businesses started hiring more than previously

The list goes on and on.  

Any one of the reasons above could have been the reason why we have seen the increase in the important metrics rather than the chatbot.
More generally the problem with observational studies is that you cannot exclude influence of other **confounders** on the outcome of your experiments. In professional language you can't **control** for those other confounders. 

If you're not convinced by this simple explanation, I would like to refer you to [Refuted observation studies](https://experimentguide.com/refuted_observational_studies/) and the [The book of Why](https://en.wikipedia.org/wiki/The_Book_of_Why).

### How does an A/B test deal with a problem of confounding?

If you recall when splitting users into test and control groups we were using a random approach, because we are trying to make sure that test and control groups are identical or at least extremely similar to each other. In fact what we are aiming to achieve is :

*To make test group different from control group in 1 thing only - presence or absence of the chatbot*. 

By doing this we are not trying to remove the confounders from the experiment. We are helpless against the confounders, but we are not helpless in creating extremely similar groups that have 1 difference only - the chatbot we are trying to test. Under those conditions : when the groups are identical to each other in everything except presence or absence of a chatbot, if we observe difference between the groups, we can conclude that it's *because of the chatbot*. In other words - A/B tests are the only reliable tool available to us that can help us infer **causality**.

We together with the rest of the tech industry were convinced that A/B test is the right way to go and we had the technology to make it happen. Hence we ran an A/B tests.









### HiPPOs 🦛
There's another way to make rollouts that follows the
"Highest paid person in the meeting" **HiPPO** system. This system is rather simple - the highest paid person in the room will tell everyone what to do. The advantage of this system is that we can do things *fast*. The disadvantage of this system is that we *rarely have an idea what was the effect of what we did fast*. \
This system can be in place under 2 different scenarios :  

*   A very young company that simply doesn't have the resouces to do A/B testing. Hopefully when a bit more resources are there A/B testing becomes part of the tools being used.
*   An established company that is not **data driven**. Instead it's driven by the gut feelings of the various HiPPos 🦛.

Hippos typically are impatient, have a big mouth and tiny ears. They are not very easy to reason with and most of the time you're going to be listening to them. They are part of the real world and you will undoubtedly encounter them in your career. To encourage you, in your interaction with them I would like to point out that HiPPos mostly eat plant based food, so you shouldn't be too afraid of their large open mouths. ## Website tracking


To understand the A/B test results we need to discuss how users were tracked.

The users were tracked using Google Analytics ('ga' for short) - a system that registers **events** that get triggered when user interacts with the website. In our case we had the following events configured :  

*   "experiment_viewed_ga_first_timestamp" - The event that gets registered when a user is assigned to either test or control group. In other words every user who was part of the experiment will have this timestamp set
*   "experiment_ga" - Which experiment was the user part of. The expected value here is 'chatbot', but there could have been other A/B tests before.
*   "variation_ga" - 'false' for chatbot disabled and 'true' for chatbot enabled.
*   "device_category_ga" - Which device the user used to access the website.
*   "application_started_first_timestamp" - The event that gets registered the first time when a user is clicking on the 'Start now' button to open the application form.
*   "form_interaction_first_timestamp" - The event that gets registered the first time when a user is entering some details in the application form.
*   "application_first_timestamp" - The event that gets registered the first time when a user is submitting the application form.


### Customer relationship management (CRM) system tracking
Once a user submits their application form they are registered in the customer relationship management system called **"Hubspot"**. The following tracking events were configred

*   "lead_first_timestamp" - The event that gets registered the first time when a user is submitting the application form. In theory it should be exactly the same as the "application_first_timestamp" in Website tracking section. In practice there are discrepancies between the 2. This is normal, because 2 different systems are used and various bugs prevent those 2 systems to be in complete sync. Those discrepancies are very common in the real world, so it's good for you to get used to it.
*   "interview_first_timestamp" - The event that gets registered the first time when a user has been scheduled an interview with the admissions team.
*   "admitted_first_timestamp" - The event that gets registered the first time when a user is admitted to the bootcamp. This happens if the interview went well and the user could pass the admission test.
*   "financing_confirmed_first_timestamp" - The event that gets registered the first time when a user has secured funding to the bootcamp.

## Simplifying assumptions to make the datatset smaller
The main question that needs to be answered for each even is : if a user had multiple interactions with the application form or with any other event is this information of vital importance and should it be present in the final dataset. For the sake of evaluating the result of this A/B test - is it important to know :

1.   That the user clicked on aplication form page
2.   That the user clicked on application form page 17 times (random number, you can substitute your own)
3.   Both

A reasonable and simplifying assumption is to say that : \
*We care whether the user performed a certain action on the website or not and we do not care how many times the user did it.*

It is true that some information will be lost, but for the sake of the current analysis we can probably live with that. If you disagree you are welcome to conduct the analysis with another simplifying assumption or without making any simplifying assumptions at all - just be clear about it to yourself and others.

If this assumption seems reasonable we can aim for a final row count of 67000 by selecting 1 of each of the multiple events for the user. We could for example look at the *first* time the user clicked on the application form, the *first* time the user interacted with it and the *first* time the user submitted it. Or we could look at the *median* interaction with a certain event or at the *last* interaction with a certain event.

# Evaluating the A/B test

We have finally reached a stage where we can start evaluating the results of our A/B test. How exactly are we going to do that? \
Let's state the business question we are trying to answer: 
**"Did the chatbot improve user experience and did it help the company make more money?".** 

By the very statement of the question you see that we are trying to achieve 2 separate goals :
1.   Improve user experience
2.   Make more money

In ideal world those goals are 100% correlated to each other - better user experience leads to making more money. But I hope that by now you probably agree that the real world is more complex than the ideal version of the world. 

It is possible, in fact not that uncommon that better user experience *does not* result in making more money. Even worse, sometimes better user experience ends up making us loose money. That's how tough real world can be 😞.

##Better user experience
It is a sad, but an empirical fact that we are very bad at understanding what our users actually want. If we weren't - conducting A/B tests would be completely unnecessary. We would just imagine how certain thing should work and it would work perfectly from the initial rollout. Life would be easy and simple. On the other hand there would be no need for data professionals to help design and evaluate A/B tests 🤓. 

First thing to note is that there is likely to be a high degree of diversity among our users and when we try to rollout a new product it is inevitable that some will hate the feature, some will be indifferent and some will like it. Better user experience is what works best for the majority - the rights of the minority are not respected, unless this minority are the ones that pay us money.  

More often than not initial rollout of a new feature or a product is not making user experience better. This is despite the best efforts of everybody involved and is an empirical fact of life. If you don't believe me, please get yourself a copy of
[Trustworthy online controlled experiments](https://www.cambridge.org/core/books/trustworthy-online-controlled-experiments/D97B26382EB0EB2DC2019A7A7B518F59).

















##Making more money
This is probably the ultimate goal of every business. The only way for a business to survive and stay competitive in a fast changing world is to innovate. It is probably possible to have a business that doesn't grow and has a stable unchanging income, that somehow has a secure position in the marketplace - a local bakery comes to mind. However it's unlikely you will be working for such a busines having completed a data bootcamp. Fundamentally if a business stops innovating, it starts loosing out to competition, stops being able to pay for itself and disappears. Those are the tough laws of life in a free market economy 🐅.

All this said - the primary goal of most product rollouts is to make more money, even if it means making user experience worse. At the very least we want to see that the new product is not causing us to loose money. ## Computing conversion rates
The first and most intuitive thing to evaluate the A/B test is to compute **conversion rates**.

Below you can see the table we have generated before.

|variation\_ga|experiment\_viewed\_ga\_first\_binary|experiment\_assignment\_hubspot\_binary|application\_started\_first\_binary|form\_interaction\_first\_binary|application\_first\_binary|lead\_first\_binary|interview\_first\_binary|admitted\_first\_binary|financing\_confirmed\_first\_binary|
|---|---|---|---|---|---|---|---|---|---|
|false|32984|1337|11613|6562|1205|1364|121|55|12|
|true|33056|1236|9738|5659|1147|1259|124|60|9| 

For example :

*   For the control version (variation_ga = false) there were 32984 users and 11613 clicked on the application started form. The proportion can be computed as 11613/32984 =  0.352. \
In other words 35.2% of users clicked the start application form button.
*   For the test version (variation_ga = true) there were 33056 users and 9738 clicked on the application started form. The proportion can be computed as 9738/33056 =  0.294. \
In other words 29.5% of users clicked the start application form button.

## Why do we need statistics?

We have computed 'application_started' conversion rate and can easily compute other conversion rates. It seems that our job here is done. I'm sad to disappoint you and tell you it's not, but don't fret - we're almost there 🙏.\
Unfortunately there's a little problem with merely computing proportions or conversion_rates - how can we be sure that the users that have interacted with our experiment are statistically speaking 'average' and representative users. Perhaps we were extremely unlucky and ended up getting users who passionately hate chatbots and if we only continue the experiment for a few more weeks we would get more chatbot positive users who would change the outcome of the experiment as a whole. \
To make this a little bit more concrete - what if we only had 10 people in control version and 10 people in the test version and 3 people converted in the control version and 4 did in the test. In this imaginary scenario control group has a conversion of 30% and test group has a conversion of 40%. Would we consider this sufficient to declare success? \
The naive answer is yes and you should expect to hear it often in your business life, the educated answer is no and aren't we educated 🎓? \
What we need is a procedure that will tell us how many observations we need in order to have confiedence that the experiment result we are looking at are **statistically significant** and not due to us randomly being lucky or unlucky.

Before we dive into the details of the procedure that will help us establish whether the result we are looking at is statistically significant, I recommend you first run and look at the output results and the plots.

## A/B test evaluation function

The function below is probably the most difficult thing to understand from this whole notebook.


````python
def evaluate_ab_test_with_plots(conversions_a, trials_a, conversions_b, trials_b, significance_level=0.05, title_binary='Density Plot of Binary Outcomes', title_conversion='Density Plot of Conversion Rates'):
    # Calculate conversion rates
    conversion_rate_a = conversions_a / trials_a
    conversion_rate_b = conversions_b / trials_b

    # Calculate pooled proportion
    pooled_proportion = (conversions_a + conversions_b) / (trials_a + trials_b)

    # Calculate standard error
    standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/trials_a + 1/trials_b))

    # Calculate z-score
    z_score = (conversion_rate_b - conversion_rate_a) / standard_error

    # Calculate p-value
    p_value = (1 - norm.cdf(abs(z_score))) * 2

    # Determine significance
    is_significant = p_value < significance_level

    # Simulate distributions of binary outcomes for plotting
    np.random.seed(0)  # For reproducibility
    dist_a = np.random.binomial(1, conversion_rate_a, trials_a)
    dist_b = np.random.binomial(1, conversion_rate_b, trials_b)

    # Simulate distributions of conversion rates for plotting
    sample_proportions_a = np.random.normal(conversion_rate_a, np.sqrt(conversion_rate_a * (1 - conversion_rate_a) / trials_a), 1000)
    sample_proportions_b = np.random.normal(conversion_rate_b, np.sqrt(conversion_rate_b * (1 - conversion_rate_b) / trials_b), 1000)

    # Plot density plots
    plt.figure(figsize=(12, 6))

    # Plot density plot of binary outcomes
    plt.subplot(1, 2, 1)
    sns.kdeplot(dist_a, label='Control Group (A)', fill=True)
    sns.kdeplot(dist_b, label='Treatment Group (B)', fill=True)
    plt.title(title_binary)
    plt.xlabel('Outcome')
    plt.ylabel('Density')
    plt.legend()

    # Plot density plot of conversion rates
    plt.subplot(1, 2, 2)
    sns.kdeplot(sample_proportions_a, label='Control Group (A)', fill=True)
    sns.kdeplot(sample_proportions_b, label='Treatment Group (B)', fill=True)
    plt.title(title_conversion)
    plt.xlabel('Conversion Rate')
    plt.ylabel('Density')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return {
        'conversion_rate_a': conversion_rate_a,
        'conversion_rate_b': conversion_rate_b,
        'z_score': z_score,
        'p_value': p_value,
        'is_significant': is_significant
    }
````

### Numerical evaluation
Let's try to understand what's happening in the function above step by step.

The function accepts the following arguments:

*   conversions_a - number of times someone in group A (control) performed a certain event)
*   trials_a - total number of users in group A (control)
*   conversions_a - number of times someone in group B (test) performed a certain event)
*   trials_b - total number of users in group B (test)
*   significance_level - the probability for us to detect a difference between control and test groups without there actually being one. If this happens in professional language it's called **Type 1 Error**. Can we ever be 100% sure that there's difference between control and test groups? But setting significance level to 0.05 means that there's a 5% chance of that happening. In business we can probably even live with 0.1, whereas in medical studies we need 0.01 significance level.
We will be using the same function to plot the differences between the various plots and the 2 arguments below will enable us to pass the appropriate plot names.
*   title_binary - title of the plot of Binary outcomes we're going to produce. We will explain what this plot shows when we go through the function.
*   title_conversion - title of the plot for Density Plot of Conversion Rates
We will explain what this plot shows when we go through the function.

Now we can start going through the code and understanding what it does :

```
    # Calculate conversion rates
    conversion_rate_a = conversions_a / trials_a
    conversion_rate_b = conversions_b / trials_b
```

This block should be familiar - we start buy calculating the conversion rates of control and test groups

```
    # Calculate pooled proportion
    pooled_proportion = (conversions_a + conversions_b) / (trials_a + trials_b)
```

Pooled proportion represents the overall conversions from both control and test group. This is the combined estimate of the proportion of successes across both groups.

```
    # Calculate standard error
    standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/trials_a + 1/trials_b))
```

Variance of a Proportion: The term ```pooled_proportion * (1 - pooled_proportion) ``` represents the variance of a binomial proportion. It is maximized when the proportion is 0.5 and minimized as the proportion approaches 0 or 1.

Sample Sizes ```(1/trials_a + 1/trials_b))``` : These terms adjust for the number of trials in each group. Larger sample sizes reduce the standard error, making the estimate more precise.

To summarise - the standard error is a measure of how much the sample statistic varies from the true population proportion. It quantifies the expected variability in the difference between the conversion rates due to random chance, i.e. sampling error. In simple words standard error estimates how much the data we observe in our sample varies from the true population. 

```
    # Calculate z-score
    z_score = (conversion_rate_b - conversion_rate_a) / standard_error
```
In the nominator we see the actual observed difference between conversion rates. In the denominator we see the the quantified expected variability in the difference between the conversion rates due to random chance. Z_score tells us how many times the observed difference is compared to expected variability in terms of standard deviations. If z_score is large in absolute terms we are looking at a difference that is unlikely to due to random chance. Formally :  

For a 95% confidence level, the critical z-score is ±1.96.
If z > 1.96 or z < -1.96, we reject the null hypothesis (i.e., the difference is statistically significant).
If z is between -1.96 and 1.96, we fail to reject the null hypothesis (i.e., the difference could be due to chance).

In the context of comparing two proportions, the z_score helps determine whether the difference between the two proportions is statistically significant.

Before we go to the next step, please note that z_score is normally distributed due to it being a calculation of conversion rates. Check the [Central limit theorem](https://en.wikipedia.org/wiki/Central_limit_theorem) if that is a surprise. 

```
    # Calculate p-value
    p_value = (1 - norm.cdf(abs(z_score))) * 2
```

p_value and z_score tell us the exact same thing, but p_value does so on a more familiar scale of 0 to 1. To perform thr transformation from z_score unbounded scale to 0-1 scale used for p_value we compute the probability that z_score is less than a certain value using ```norm.cdf```. This function calculates the cumulative distribution function (CDF) for a normal distribution and for our z_score tells us what is the probability of obtaining a value smaller than the z_score calculated. Conversely ```(1 - norm.cdf(abs(z_score)))``` gives us the probability of obtaining a value greater than z. Multiplying by 2 is a reflection of the fact that we are checking whether the conversion rate of one group is significantly greater or smaller than the other group. If we were checking the change in just 1 direction we would not need to multiply by 2. 

So p_value is telling us how likely it is that the difference between the conversion rates of the groups we have observed is due to random chance on a scale of 0 to 1. To make a decision whether we trust the difference in conversion rates observed to be *not due to random chance* all we need now to do is agree on a threshold of probability above which we do not trust the result. This threshold is called "significance levels". In our case we are using a significance level of 0.05, but we could be using other significance levels - a significance level of 0.1 for example would accept a 10% chance of concluding we are seeing a difference between conversion rates where there is no difference and a significance level of 0.001 would only accept a 1% chance of concluding we are seeing a difference in conversion rates where there is no difference. 
0.05 is a standard significance level that is used in most non-medical research and industry, so we are using it as well.


```
    # Determine significance
    is_significant = p_value < significance_level
```
Finally we compare our p_value with the chosen significance level and output a conclusion
